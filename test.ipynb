```python
import requests
from bs4 import BeautifulSoup
```

```python
pdfs_set = set()
url_base = 'https://www.enedis.fr/documents'
url = url_base
while True:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    recoverd_pdf = soup.select('.pdf')
    for pdf in recoverd_pdf:
        pdfs_set.add("https://www.enedis.fr" + pdf.attrs.get("href"))
    next = soup.select(".pager__next")
    print(len(pdfs_set))
    if not next:
        break
    url = url_base + next[0].attrs.get("href")
```

```python
data = []
url_visited = []
url_base = 'https://www.enedis.fr/documents'
url = url_base

for num_page in range(10):
    print(num_page)
    response = requests.get(url)
    url_visited.append(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    scrapping = []
    page_content = soup.select(".press-page__field")
    nb_elem_to_get = len(page_content) // 6

    for i in range(nb_elem_to_get):
        elem = []
        for j in range(6):
            elem.append([elem for elem in page_content[i * 6 + j].contents if elem != '\n'])
        elem.remove([])
        for k in range(len(elem)):
            if elem[k]:
                elem[k] = elem[k][0]
            else:
                elem[k] = ""
        scrapping.append(elem)
        
    for pdf in scrapping:
        elem = []
        elem.append("https://www.enedis.fr" + pdf[0].attrs.get('href'))
        elem.append(str(pdf[1]))
        elem.append(pdf[2].__dict__.get("attrs", {'datetime':pdf[2]})['datetime'])
        elem.append(str(pdf[3]))
        elem.append(pdf[4].__dict__.get("attrs", {'aria-label':pdf[4]})['aria-label'])
        data.append(elem)

    next = soup.select(".pager__next")
    url = url_base + next[0].attrs.get("href")
```
